```python
# Gráfico de dispersión
sns.scatterplot(x='fecha_matrimonio', y='comuna', data=Unioncivil_2019)
plt.title('Diagrama de Dispersión')
plt.show()
```

```python
# Asegurarse de que la columna 'fecha_union' esté en formato datetime
Unioncivil_2019['fecha_matrimonio'] = pd.to_datetime(Unioncivil_2019['fecha_matrimonio'], errors='coerce')
```

```python
# Crear una nueva columna 'mes_anio' que contenga el mes y año de 'fecha_union'
Unioncivil_2019['mes_anio'] = Unioncivil_2019['fecha_matrimonio'].dt.to_period('M').astype(str)
```

```python
# Eliminar filas con valores nulos en las columnas relevantes
Unioncivil_2019 = Unioncivil_2019.dropna(subset=['mes_anio', 'comuna'])
```

```python
# Crear el gráfico de dispersión
sns.scatterplot(x='mes_anio', y='comuna', data=Unioncivil_2019)
plt.title('Diagrama de Dispersión por Mes y Comuna')
plt.xticks(rotation=45)  # Rota las etiquetas del eje X para mejor visualización
plt.show()
```
